```python
from tkinter import *
import sqlite3
import tkinter.messagebox

root = Tk()
root.geometry('500x500')
root.resizable(0, 0)
root.title("Registration")


fullname=StringVar()
email=StringVar()
gender = StringVar()
dept=StringVar()
empID= StringVar()



def Register():
    Lname=fullname.get()
    Lemail=email.get()
    Lgender=gender.get()
    Ldept=dept.get()
    LempID=empID.get()
    conn = sqlite3.connect('OperatorSafetyGear.db')
    with conn:
        cursor=conn.cursor()
        cursor.execute('CREATE TABLE IF NOT EXISTS Employee_details(Fullname TEXT,Email TEXT,Gender TEXT,Department TEXT ,EmployeeID TEXT)')
        cursor.execute('INSERT INTO Employee_details(FullName,Email,Gender,Department,EmployeeID) VALUES(?,?,?,?,?)',(Lname,Lemail,Lgender,Ldept,LempID))
        conn.commit()        
        tkinter.messagebox.showinfo("Registration", "Successful")   
             
label_0 = Label(root, text="Registration",width=20,font=("bold", 20))
label_0.place(x=90,y=53)


label_1 = Label(root, text="FullName",width=20,font=("bold", 10))
label_1.place(x=80,y=130)

entry_1 = Entry(root,textvar=fullname)
entry_1.place(x=240,y=130)

label_2 = Label(root, text="Email",width=20,font=("bold", 10))
label_2.place(x=68,y=180)

entry_2 = Entry(root,textvar=email)
entry_2.place(x=240,y=180)

label_3 = Label(root, text="Gender",width=20,font=("bold", 10))
label_3.place(x=70,y=230)

Radiobutton(root, text="Male",padx = 5, variable=gender, value="M").place(x=235,y=230)
Radiobutton(root, text="Female",padx = 20, variable=gender, value="F").place(x=290,y=230)
gender.set('M')


label_4 = Label(root, text="Department",width=20,font=("bold", 10))
label_4.place(x=70,y=280)
list1 = ['A','B','C','D'];
droplist=OptionMenu(root,dept, *list1)
droplist.config(width=15)
dept.set('Select Department')
droplist.place(x=240,y=280)

label_5 = Label(root, text="Employee ID",width=20,font=("bold", 10))
label_5.place(x=85,y=330)

entry_5 = Entry(root,textvar=empID)
entry_5.place(x=235,y=330)


Button(root, text='Register',width=20,bg='brown',fg='white',command=Register).place(x=180,y=380)


root.mainloop()
```
